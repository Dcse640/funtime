```python
import pandas as pd

# Read the Titanic dataset
df = pd.read_csv('titanic.csv')

# Total number of passengers in each class
total_first_class = df[df['Pclass'] == 1].shape[0]
total_second_class = df[df['Pclass'] == 2].shape[0]
total_third_class = df[df['Pclass'] == 3].shape[0]

# Filter the dataset for survivors in each class
survived_first_class = df[(df['Pclass'] == 1) & (df['Survived'] == 1)]
survived_second_class = df[(df['Pclass'] == 2) & (df['Survived'] == 1)]
survived_third_class = df[(df['Pclass'] == 3) & (df['Survived'] == 1)]

# Count the number of survivors in each class
num_survived_first_class = survived_first_class.shape[0]
num_survived_second_class = survived_second_class.shape[0]
num_survived_third_class = survived_third_class.shape[0]

# Calculate survival percentages
if total_first_class > 0:
    percent_survived_first_class = (num_survived_first_class / total_first_class) * 100
else:
    percent_survived_first_class = 0

if total_second_class > 0:
    percent_survived_second_class = (num_survived_second_class / total_second_class) * 100
else:
    percent_survived_second_class = 0

if total_third_class > 0:
    percent_survived_third_class = (num_survived_third_class / total_third_class) * 100
else:
    percent_survived_third_class = 0

# Display the counts, percentages, and details for each class
print(f"Number of 1st class survivors: {num_survived_first_class}")
print(f"Survival percentage for 1st class: {percent_survived_first_class:.2f}%")
print(survived_first_class[['Name', 'Pclass', 'Age', 'Fare']])
print()

print(f"Number of 2nd class survivors: {num_survived_second_class}")
print(f"Survival percentage for 2nd class: {percent_survived_second_class:.2f}%")
print(survived_second_class[['Name', 'Pclass', 'Age', 'Fare']])
print()

print(f"Number of 3rd class survivors: {num_survived_third_class}")
print(f"Survival percentage for 3rd class: {percent_survived_third_class:.2f}%")
print(survived_third_class[['Name', 'Pclass', 'Age', 'Fare']])

```
